```python
import tkinter as tk
from tkinter import messagebox

class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        self.balance += amount
        self.transactions.append(f"Deposit: +${amount}")

    def withdraw(self, amount):
        if amount > self.balance:
            messagebox.showerror("Error", "Insufficient funds")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawal: -${amount}")

class OnlineBankingSystem:
    def __init__(self, master):
        self.master = master
        self.master.title("Online Banking System")
        self.master.geometry("400x300")

        self.label_name = tk.Label(master, text="Username:")
        self.label_name.pack()
        self.entry_name = tk.Entry(master)
        self.entry_name.pack()

        self.label_password = tk.Label(master, text="Password:")
        self.label_password.pack()
        self.entry_password = tk.Entry(master, show="*")
        self.entry_password.pack()

        self.button_login = tk.Button(master, text="Login", command=self.login)
        self.button_login.pack()

    def login(self):
        username = self.entry_name.get()
        password = self.entry_password.get()

        # Dummy authentication, replace with your actual authentication logic
        if username == "shreya" and password == "1319":
            self.show_account_interface()
        else:
            messagebox.showerror("Error", "Invalid username or password")

    def show_account_interface(self):
        self.master.destroy()

        account_window = tk.Tk()
        account_window.title("Account Management")

        account = BankAccount("User")

        def deposit():
            amount = float(entry_deposit.get())
            account.deposit(amount)
            messagebox.showinfo("Success", f"${amount} deposited successfully")

        def withdraw():
            amount = float(entry_withdraw.get())
            account.withdraw(amount)
            messagebox.showinfo("Success", f"${amount} withdrawn successfully")

        def show_balance():
            messagebox.showinfo("Balance", f"Current Balance: ${account.balance}")

        def show_transactions():
            transaction_text = "\n".join(account.transactions)
            messagebox.showinfo("Transaction History", transaction_text)

        label_deposit = tk.Label(account_window, text="Deposit Amount:")
        label_deposit.grid(row=0, column=0)
        entry_deposit = tk.Entry(account_window)
        entry_deposit.grid(row=0, column=1)
        button_deposit = tk.Button(account_window, text="Deposit", command=deposit)
        button_deposit.grid(row=0, column=2)

        label_withdraw = tk.Label(account_window, text="Withdraw Amount:")
        label_withdraw.grid(row=1, column=0)
        entry_withdraw = tk.Entry(account_window)
        entry_withdraw.grid(row=1, column=1)
        button_withdraw = tk.Button(account_window, text="Withdraw", command=withdraw)
        button_withdraw.grid(row=1, column=2)

        button_balance = tk.Button(account_window, text="Check Balance", command=show_balance)
        button_balance.grid(row=2, column=0)

        button_transactions = tk.Button(account_window, text="Transaction History", command=show_transactions)
        button_transactions.grid(row=2, column=1)

        account_window.mainloop()

def main():
    root = tk.Tk()
    app = OnlineBankingSystem(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
